```python
#print numbers 1-10
print("for loop")
for number in range(1, 11):
    print(number)
```

```python
#print numbers 1 - 10 but break at certain conditions

print("Break loop")
for number in range(1, 11):
    if number == 5:
        break
    print(number)
```

```python
#print numbers from  1 to 10 but continue loop under certain conditions
```

```python
print("continue loop")
for number in range(1,11):
    if number % 2: #modulus operator
        continue #next iteration
    print(number)
```

```python
print("Iterate loop")
items = ["Books", "Banana", "Computer"] #list
for item in items:
    print("I will take my ", item, "with me")
```

```python
print("While loop")
number = 1
while(number <= 10):
    print(number)
    number += 1
```

```python
#Read a string from the user, and Print characters present at an even index number

userinput = input("What is you name? ")

for i in range(len(userinput)):
    if i % 2 == 0:
        print(userinput[i])
```

```python
# # 2 - Write a Python code to remove characters from a string from 0 to n and print the new string
# Hint, Use string slicing to get a substring. For example, remove the first four characters using s[4:]

name: str = "Kenya "

print(name[0:2])

print(name[::3])

print(name[:3])
```

```python
#3 - Find the number of occurrences of a substring in a string
# Hint, Use string method count(). E.g., s.count("something")

sentence = "My home is Kenya. Kenya is a beautiful country in East Africa. It is the home of World " \
"Champions in Athletics"

print(sentence.count("Kenya")) #gives a count of the number of occurences
print(sentence.find("My")) #gives the index(position) of the string
```

```python
#4 - Write a Python code to check if the given number is palindrome. 
# A palindrome number is a number that is the same after reverse. 
# For example, 545 is the palindrome number.
# Hint,Reverse the given number and save it in a different variable. 
# Use the if condition to check if the original and reverse numbers are identical. 

number = int(input("Enter a number: "))
number1 = number.reverse()

if number == number1:
    print('Palindrome Number')
```
