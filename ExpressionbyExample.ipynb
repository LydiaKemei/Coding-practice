Basic number Handling

```python
i = 1
f = 2.5

print("Numbers: ", i, ",", f) #print numbers
```

```python
print("Add", i + f) #add numbers
```

```python
print("Divide ", f/2) #divide numbers
```

```python
print("i2f: ", float(i)) #convert int to float
```

```python
print("f2i: ", int(f)) #convert float to int
```

```python
print("round: ", int(round(f))) #Do proper rounding during conversion
```

```python
print("Modulo: ", 5%2) # Modulo operation - used to find the remainder after dividing one number by another.
```

Work with strings

```python
s = "This is my string with some meaningful text"
```

```python
print(s)
```

```python
print(s.split(" ")) #separate by spaces
```

```python
print(s.replace("s", "x")) #Replace characters
```

```python
print(s.lower()) #To lower case
```

```python
print(s.upper()) #To upper case
```

```python
print(s.find("m")) #find a character in the string
```

ACCESSING PART OF A STRING (DATA STRUCTURE)

```python
print(s[:4]) #print part of the string, all characters till the 4th
```

```python
print(s[1:4]) #2nd to 4th character
```

```python
print(s[-4:]) #Last four characters
```

```python
print(s)
```

```python
print(s[::2]) #Every 2nd character
```

Converting between different types

```python
floatString = "12.3"
intString = "45"

print("Convert and add 100 to the number")
```

```python
print("Convert string", intString, "to float ", float(intString)+100)
```

```python
print("Convert string", floatString, "to float: ", float(floatString)+100)
```

```python
print("Convert string", intString, "to int: ", int(intString)+100)
```

```python
print("Convert a number to a string: ")
print("Number: ", str(123.54) + str(100))
```

Work with tuple, set and lists

```python
a = [1, 2, 3, 4]
b = [5, 6, 7, 8]
# unpack trick
c = (a, b)
print(c)
```

```python
#unpacking the list into individual elements
c = (*a, *b)
print(c)
```
